```python
import pandas as pd
```

```python
df = pd.read_excel("2018-2019 Stop Data PART 1.xlsx")
df2 = pd.read_excel("2018-2019 Stop Data PART 2.xlsx")
```

```python
df_merge = [df, df2]
df_main = pd.concat(df_merge)
df_main
```

```python
#dataframe with only the green route
green = df_main[df_main.Route == 'Green']
```

```python
#this will be the main dataframe that I use
main = green[['Date', 'Count', 'On off', 'Stop']]
```

```python
main
```

```python
#Will use these values
main.Stop.unique().size
```

```python
main.dropna()
```

```python
main = main.rename(columns={"On off": "Entry"})

#These will be the frames for the on data and the off data
main_on = main[main.Entry == 'on']
main_off = main[main.Entry == 'off']
```

```python
main_on.reset_index()
main_off.reset_index()
main.reset_index()
```

```python
#MAKE SURE TO RUN THIS ENTIRE BLOCK TOGETHER TO RESET THE CALC[] ARRAY VALUES.

#This array will be used to iterate through in a for loop
stopsArray = main.Stop.unique()

#This array will store the average values of every stop
calc = []

#used to calculate averages
overallStops = 0

#Ran this 3 separate times - once with the main_on, main_off, and main array to obtain the different values.
for x in stopsArray:
    for index, row in main_off.iterrows():
        if row['Stop'] == x:
            overallStops = overallStops + row['Count']
    calc.append(overallStops)
    overallStops = 0
```

```python
#This returned the values and I matched them with the stops I obtained from the overallStops variable.
calc
```
