```python
#group data in hour interval and count the number of people in bus
df_on = df[df.Onoff =='on']
df_on['ts'] = pd.to_datetime(df_on['Date'] + ' ' + df_on['Time'])
interval = df_on.set_index('ts')['Count'].resample('1H').sum()
interval
```

```python
#divide data into 2 semesters Fall and Spring
df_interval = interval.to_frame()
fall = df_interval['2018-07':'2018-12']
spring = df_interval['2019-01':]
#drop rows where count=0
fall = fall[fall.Count !=0]
spring = spring[spring.Count !=0]
fall
```
